# **Univariate Analysis**

```python
# 2. To see the highest of the uniot price.
#Histogram
import matplotlib.pyplot as plt
import seaborn as sns

price=dataset['Unit price']

# Generate a histogram
plt.figure(figsize=(10, 6))
plt.hist(price, bins=8, color='skyblue', edgecolor='black')
plt.title("Data Visualization Scores")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

 ** Insights :-**
  
  Data Distribution: The histogram provides an overview of the distribution of unit prices in your dataset. The x-axis represents the price range, and the y-axis represents the frequency or count of data points falling into each price range.

  Center and Spread: You can observe the central tendency of the data, which is where the peak of the histogram is located. In this case, it corresponds to the most common unit price. Additionally, the spread or variation in prices can be inferred from the width of the histogram.

  Skewness: The shape of the histogram can tell you about the skewness of the data. If the histogram is skewed to the left (negatively skewed), it means that lower prices are more common. If it's skewed to the right (positively skewed), it indicates that higher prices are more common.

  Outliers: You can identify potential outliers if there are bars (bins) that extend significantly farther to the right or left of the main distribution. Outliers may represent unusual or extreme price values.

  Bin Size: The choice of the number of bins (in this case, 8) can impact the appearance of the histogram. A smaller number of bins might smooth out details, while a larger number of bins might reveal finer-grained patterns in the data.

  Data Quality: You can assess data quality by examining the histogram. For instance, if you see gaps or strange patterns in the histogram, it could be an indication of data entry errors or inconsistencies.

  Summary Statistics: While not shown in the code you provided, you can also calculate summary statistics such as the mean, median, and standard deviation to provide more numerical insights into the distribution.

# **Bivariate Analysis**

```python
# 3.What is the distribution of gross income for each product line?
# Swarm Plot
plt.figure(figsize=(12,8))
sns.swarmplot(x='Product line', y='gross income',hue="Payment", data=dataset, color='skyblue')
plt.title("swarm Plot: gross income of particular product line")
plt.xlabel("Product line'")
plt.ylabel("gross income")
plt.show()
```

 ** Insights :-**
  
  Product Line vs. Gross Income: The x-axis of the plot represents different "Product lines." Each point along the x-axis represents a unique product line category. You can see how "gross income" varies for each product line.

  Gross Income Distribution: The vertical position of the points on the y-axis represents the "gross income" for each data point. The spread and distribution of data points for each product line provide insights into the variation in gross income for different product lines. You can observe whether some product lines tend to generate higher or lower gross income.

  Payment Method: The "hue" parameter is set to "Payment," which means that the points are color-coded based on the payment method. Each color represents a different payment method. This allows you to see if the payment method has any influence on the gross income for a specific product line.

  Data Density: The swarm plot helps to visualize the density of data points. Clusters of points suggest that specific product lines or payment methods are more common, and you can observe their associated gross incomes.

  Outliers: Swarm plots are useful for identifying outliers. Any data points that are far away from the main clusters may indicate unusual or unexpected gross income values for a particular product line and payment method combination.

  Title and Labels: The title of the plot provides context, and the x and y-axis labels make it clear what is being visualized.

```python
# 4.How do gross incomes vary across different product lines for both male and female customers?
#Bar plot
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a bar plot with hue (Class and Division)
plt.figure(figsize=(12, 6))
sns.barplot(x='Product line',y="gross income", hue='Gender', data=dataset, ci=None, palette='Set2')
plt.title("Distribution of gross income with product line")
plt.xlabel("Product line")
plt.ylabel("gross income")
plt.show()
```

  Gender Differentiation: The bars are further divided into segments, each representing a different gender. In this plot, you can see how gross income varies across product lines for different genders. The color differentiation makes it easy to compare the performance of product lines and genders.

  Comparing Product Lines: You can compare the heights of the bars to see which product lines generate the highest and lowest gross income. Are there significant differences between the product lines in terms of income generation?

  Gender Analysis: By comparing the segments within each bar, you can analyze how gross income is distributed across genders within each product line. Are there gender-based differences in purchasing behavior for specific product lines?

  Distribution Shape: Pay attention to the distribution shape of the bars. Are there product lines where the gross income is evenly distributed among genders, or are there cases where one gender dominates in terms of income?

  Title and Labels: The title, x-axis label, and y-axis label provide context and make it clear what the plot is depicting.

  Palette: The chosen color palette ('Set2') is used to differentiate the genders, making it visually appealing and easier to interpret.

  No Confidence Intervals (ci=None): Setting the confidence intervals to None means that error bars showing confidence intervals are not displayed on the bars. This can make the plot cleaner.

```python
# 5.Is there correlation between the unit price , rating and total
#correlation Heatmap
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
corr_matrix = dataset[['Unit price', 'Total','Rating']].corr()

# Generate a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='viridis',linewidth=0.5, fmt=".2f")
plt.title("Correlation of Unit price ,Total and Rating",color='m')
plt.show()
```

**  Insights :-**
  
  Correlation Values: In the heatmap, the cells show correlation coefficients between pairs of variables. The values range from -1 to 1, with 1 indicating a perfect positive correlation, -1 indicating a perfect negative correlation, and 0 indicating no correlation. The values in your heatmap are annotated for clarity.

  Positive Correlation: If you see a high positive correlation (values close to 1), it means that as one variable increases, the other variable tends to increase as well. In your case, you might find that 'Total' and 'Rating' have a positive correlation. This implies that, generally, higher total amounts are associated with higher ratings.

  Negative Correlation: If you see a high negative correlation (values close to -1), it means that as one variable increases, the other variable tends to decrease. There doesn't appear to be a strong negative correlation between any of these variables in your heatmap.

  Weak or No Correlation: Values close to 0 indicate little to no linear relationship between the variables. For example, if 'Unit price' has a correlation close to 0 with 'Total' or 'Rating,' it suggests that changes in 'Unit price' do not significantly impact 'Total' or 'Rating.'

  Data Relationships: The heatmap helps you identify relationships within your data. For example, you can quickly spot which pairs of variables have the strongest correlations.

  Data Quality and Feature Selection: Correlation analysis can be used in feature selection. If you find that two variables have a very high correlation, you might consider using only one of them in your analysis to reduce multicollinearity.

  Title and Color: The title of the heatmap indicates that it shows the correlation between 'Unit price,' 'Total,' and 'Rating.' The color map used ('viridis') provides a clear representation of the correlation values, with brighter colors indicating stronger correlations.
